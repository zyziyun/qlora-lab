# 04 - Serve it: merge vs hot-swappable adapter

Two options, and the trade-off is the ML-infra interview answer:

- **merge**: fold LoRA into the base -> one standalone model. Simplest, zero
  routing latency, but one full model per task.
- **vLLM LoRA**: load the base once, hot-swap many adapters by request name.
  One GPU + N small files serves N task variants.
